# Моделирование длинной линии 
Пусть линия с волновым сопротивлением W = 50 Ом нагружена на произвольное сопротивление Zн=Rн+jXн (активное сопротивление Rн, реактивное сопротивление Хн). Вывести графики эпюр распределения амплитуд напряжения и тока вдоль линии:

lmd = 1 # длина волны для вычисления волнового числа 2pi/lmd 

W = 50 # волновое сопротивление линии 

Rn = W * 1 

Xn = W * 0.5 

Zn = Rn + Xn*1j # нормировано к W 


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
lmd = 1 
W = 50 
Rn = W * 1
Xn = W * 0.5
Zn = Rn + Xn*1j 
Un=1 
```

```python
In=Un/Zn
z=np.linspace(0,2,1000)
U=Un*np.cos(2*np.pi/lmd*z)+In*W*np.sin(2*np.pi/lmd*z)
I=In*np.cos(2*np.pi/lmd*z)+Un/W*np.sin(2*np.pi/lmd*z)
```

```python
fig1, ax = plt.subplots(figsize=(10, 5), dpi=600)
ax.plot(z,np.abs(U))
ax.grid()
ax.set_xlim([2,0])
ax.set_xlabel('z, м')
ax.set_ylabel('|U|, В')
plt.title("Распределение амплитуд напряжения")
```

```python
fig2, ax = plt.subplots(figsize=(10, 5), dpi=600)
ax.plot(z,np.abs(I))
ax.grid()
ax.set_xlim([2,0])
ax.set_xlabel('z, м')
ax.set_ylabel('|I|, А')
plt.title("Распределение амплитуд тока")
```

```python
Zvx=U/I
```

```python
fig3, ax = plt.subplots(figsize=(10, 5), dpi=600)
ax.plot(z,np.abs(Zvx))
ax.grid()
ax.set_xlim([2,0])
ax.set_ylim([35,65])
ax.set_xlabel('z, м')
ax.set_ylabel('|Zвх|, Ом')
plt.title("Входное сопротивление")
```
