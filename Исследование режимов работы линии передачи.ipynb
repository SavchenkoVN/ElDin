# Исследование режимов работы линии передачи.
## Задание:  
 Получить эпюры тока и напряжения, соответствующие всем режимам работы длинной линии.
 Для этого смоделировать работу линии на следующие виды нагрузок (Zн = Rн + jXн):  
 1. Линия, нагруженная на реактивное сопротивление (Zн = jXн)
 2. Линия, нагруженная активное сопротивление (Zн = Rн , рассмотреть случаи: Rн > W; Rн < W; Rн = W)
 3. Линия, нагруженная на комплексное сопротивление (Zн = Rн + jXн)
 
 

##  1.Линия, нагруженная на реактивное сопротивление. 
Пусть Zн = jXн; Uн=1 [В]


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
W = 1
lmd = 1
Un=1 
```

```python
def U(z,Zn):
    In=Un/Zn
    return Un*np.cos(2*np.pi/lmd*z)+In*W*np.sin(2*np.pi/lmd*z)*1j
def I(z,Zn):
    In=Un/Zn
    return In*np.cos(2*np.pi/lmd*z)+Un/W*np.sin(2*np.pi/lmd*z)*1j
z=np.linspace(0,2,1000)*lmd
Xn=1*W
Zn1=1j*Xn
U1=U(z,Zn1)
I1=I(z,Zn1)
```

```python
fig1, ax = plt.subplots(2,1,figsize=(10, 10), dpi=600)

ax[0].plot(z,np.abs(U1))
ax[0].grid()
ax[0].set_xlim([2,0])
ax[0].set_ylim([0, max(abs(U1))+1/7*max(abs(U1))])
ax[1].set_ylim([0, max(abs(I1))+1/7*max(abs(I1))])
ax[0].set_xlabel('$z/\lambda$')
ax[0].set_ylabel('|U|, В')
ax[0].set_title("Распределение амплитуд напряжения")
ax[1].set_title("Распределение амплитуд тока")
ax[1].plot(z,np.abs(I1),'r')
ax[1].grid()
ax[1].set_xlim([2,0])
ax[1].set_xlabel("$z/\lambda$")
ax[1].set_ylabel('|I|, А')
```

Расчитаем коэффициент бегущей волны и коэффициент отражения

```python
G1=((Zn1-W)/(Zn1+W))
Kbv1=(1-abs(G1))/(1+abs(G1))
print("Кбв =",Kbv1,
     "|Г| =",abs(G1))
```

В данном случае коэффициент стоячей волны равен бесконечности.
Наблюдается режим стоячей волны.

## 3.Линия, нагруженная на комплексное сопротивление (Zн = Rн + jXн).
Zн = Rн + jXн

Расчитаем коэффициент бегущей волны и коэффициент отражения:

```python
G3=((Zn3-W)/(Zn3+W))
Kbv3=(1-abs(G3))/(1+abs(G3))
Ksv3=1/Kbv3
print("Кбв =",Kbv3,
     "|Г| =",abs(G3),"Ксв=",Ksv3)
```

Наблюдается режим смешанных волн.
